**Q1.** To compute the sample mean of a variable $X$:
$$
m(X) = \dfrac{1}{N} \sum_{i=1}^N x_i
$$
and sample covariance of two variables $X$ and $Y$,
$$
\text{cov}(X,Y) = \dfrac{1}{N} \sum_{i=1}^N (x_i - m(X))(y_i - m(Y))).
$$
Recall, the sample variance of $X$ is
$$
s^2 = \dfrac{1}{N} \sum_{i=1}^N (x_i - m(X))^2.
$$

**A1. Utilized Math2Code for Equations**

1. Show that $m(a + bX) = a+b \times m(X)$.
   $$
   m(a + bX) = \frac{1}{N} \sum_{i=1}^N (a + b x_i) = \frac{1}{N} \sum_{i=1}^N a + \frac{1}{N} \sum_{i=1}^N b x_i
   $$
   $$
   = a + b \left( \frac{1}{N} \sum_{i=1}^N x_i \right) = a + b m(X)
   $$

2. Show that $\text{cov}(X,a+bY) = b \times \text{cov}(X,Y)$
   $$
   \text{cov}(X, a+bY) = \frac{1}{N} \sum (x_i - m(X))((a+by_i) - m(a+bY))
   $$
   $$
   = \frac{1}{N} \sum (x_i - m(X))(a+by_i - (a+b m(Y)))
   $$
   $$
   = \frac{1}{N} \sum (x_i - m(X))(b(y_i - m(Y)))
   $$
   $$
   = b \frac{1}{N} \sum (x_i - m(X))(y_i - m(Y)) = b \text{cov}(X, Y)
   $$

3. Show that $\text{cov}(a+bX,a+bX) = b^2 \text{cov}(X,X) $, and in particular that $\text{cov}(X,X) = s^2 $.
   Using the property from part 2:
   $$
   \text{cov}(a+bX, a+bX) = b \text{cov}(a+bX, X)
   $$
   Using symmetry of covariance $\text{cov}(A,B) = \text{cov}(B,A)$:
   $$
   = b \text{cov}(X, a+bX)
   $$
   Using property 2 again:
   $$
   = b (b \text{cov}(X, X)) = b^2 \text{cov}(X, X)
   $$
   Since $\text{cov}(X,X) = \frac{1}{N} \sum (x_i - m(X))(x_i - m(X)) = \frac{1}{N} \sum (x_i - m(X))^2 = s^2$, we have $b^2 s^2$.

4. For the median:
   Yes, for a monotonically increasing transformation $g(x)$, the median of the transformed variable is the transformed median. This applies to any quantile.
   However, it does not apply to the IQR or Range in a simple multiplicative way unless the transformation is linear.

5. Mean of transformation:
   No, generally $m(g(X)) \neq g(m(X))$ for non-linear $g$. This is known as Jensen's Inequality.

**Q2.** This question looks at financial transfers from foreign entities to American universities. In particular, from which countries and giftors are the gifts coming from, and to which institutions are they going? For this question, `.groupby([vars]).count()` and `.groupby([vars]).sum()` will be especially useful to tally the number of occurrences and sum the values of those occurrences.

1. Load the `./data/ForeignGifts_edu.csv` dataset.
2. For `Foreign Gift Amount`, create a histogram and describe the variable. Describe your findings.
3. For `Gift Type`, create a histogram or value counts table. What proportion of the gifts are contracts, real estate, and monetary gifts?
4. Create a kernel density plot of the log of `Foreign Gift Amount`, and then a kernel density plot of the log of `Foreign Gift Amount` conditional on gift type. Do you notice any patterns?
5. What are the top 15 countries in terms of the number of gifts? What are the top 15 countries in terms of the amount given?
6. What are the top 15 institutions in terms of the total amount of money they receive? Make a histogram of the total amount received by all institutions. 
7. Which giftors provide the most money, in total?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#1. Load data
df_foreign = pd.read_csv('./data/ForeignGifts_edu.csv')

#2. Foreign Gift Amount histogram
plt.figure()
df_foreign['Foreign Gift Amount'].hist(bins=50)
plt.title('Distribution of Foreign Gift Amounts')
plt.show()
print(df_foreign['Foreign Gift Amount'].describe())

#3. Gift Type
print(df_foreign['Gift Type'].value_counts(normalize=True))
df_foreign['Gift Type'].value_counts().plot(kind='bar')
plt.title('Gift Type Distribution')
plt.show()

#4. Kernel density
df_foreign['log_amount'] = np.log(df_foreign['Foreign Gift Amount'] + 1)
sns.kdeplot(df_foreign['log_amount'])
plt.title('Log Foreign Gift Amount KDE')
plt.show()

sns.kdeplot(data=df_foreign, x='log_amount', hue='Gift Type', common_norm=False)
plt.title('Log Gift Amount by Type')
plt.show()

#5. Top 15 countries
print("Top 15 Countries by Count:")
print(df_foreign.groupby('Country of Giftor')['Country of Giftor'].count().sort_values(ascending=False).head(15))
print("\nTop 15 Countries by Amount:")
print(df_foreign.groupby('Country of Giftor')['Foreign Gift Amount'].sum().sort_values(ascending=False).head(15))

#6. Top 15 institutions
print("\nTop 15 Institutions by Amount:")
print(df_foreign.groupby('Institution Name')['Foreign Gift Amount'].sum().sort_values(ascending=False).head(15))
df_total_inst = df_foreign.groupby('Institution Name')['Foreign Gift Amount'].sum()
plt.figure()
df_total_inst.hist(bins=50)
plt.title('Total Amount Received by Institutions')
plt.show()

#7. Top giftors
print("\nTop Giftors by Amount:")
print(df_foreign.groupby('Giftor Name')['Foreign Gift Amount'].sum().sort_values(ascending=False).head(5))
```

4. Instead of levels, we want to look at change over time. Compute the first difference of  `Active Well Service Rig Count (Number of Rigs)` and plot it over time. Describe what you see.

```python
#1. Load data
df_rigs = pd.read_csv('./data/drilling_rigs.csv')
print(df_rigs.info())

#2. Date conversion
df_rigs['time'] = pd.to_datetime(df_rigs['Month'], format='mixed')

#3. Line plot
plt.figure()
plt.plot(df_rigs['time'], df_rigs['Active Well Service Rig Count (Number of Rigs)'])
plt.title('Active Rigs Over Time')
plt.show()

#4. First difference
df_rigs['diff_rigs'] = df_rigs['Active Well Service Rig Count (Number of Rigs)'].diff()
plt.figure()
plt.plot(df_rigs['time'], df_rigs['diff_rigs'])
plt.title('Change in Active Rigs')
plt.show()

#5. Melt
df_melt = df_rigs.melt(id_vars=['time'], value_vars=['Onshore Rig Count (Number of Rigs)', 'Offshore Rig Count (Number of Rigs)'], var_name='Rig Type', value_name='Count')
sns.lineplot(data=df_melt, x='time', y='Count', hue='Rig Type')
plt.show()
```
